### Data Science Job Salaries Dataset contains 11 columns, each are:

- work_year: The year the salary was paid.
- experience_level: The experience level in the job during the year
- employment_type: The type of employment for the role
- job_title: The role worked in during the year.
- salary: The total gross salary amount paid.
- salary_currency: The currency of the salary paid as an ISO 4217 currency code.
- salaryinusd: The salary in USD
- employee_residence: Employee's primary country of residence in during the work year as an ISO 3166 country code.
- remote_ratio: The overall amount of work done remotely
- company_location: The country of the employer's main office or contracting branch
- company_size: The median number of people that worked for the company during the year

```python
#Which countries have extreme salary differences (large IQR or many outliers)?
```

```python
# Calculating IQR for each country
salary_summary = dss.groupby("employee_residence")["salary_in_usd"].describe()
salary_summary["IQR"] = salary_summary["75%"] - salary_summary["25%"]

# Sort by largest IQR (salary variation)
top_salary_variation = salary_summary.sort_values(by="IQR", ascending=False).head(5)

print("Countries with the Largest Salary Variation (IQR):")
print(top_salary_variation[["IQR"]])
```

```python
#Which countries have the most extreme high salaries (outliers)?
```

```python
# Sorting by highest max salaries (outliers)
top_outliers = salary_summary.sort_values(by="max", ascending=False).head(5)

print("\nCountries with the Most Extreme High Salaries (Outliers):")
print(top_outliers[["max"]])
```

## Comparison of remote vs. On-Site Salaries

```python
# Calculate average salary based on remote work ratio
avg_salary_by_remote = dss.groupby("remote_ratio")["salary_in_usd"].mean()

# Extract values for on-site (0%), hybrid (50%), and fully remote (100%) roles
on_site_salary = avg_salary_by_remote.get(0)
hybrid_salary = avg_salary_by_remote.get(50)
remote_salary = avg_salary_by_remote.get(100)

# Print results
print("Average Salary by Remote Work Ratio:")
print(f"On-Site (0% Remote): ${on_site_salary:,.0f}")
print(f"Hybrid (50% Remote): ${hybrid_salary:,.0f}")
print(f"Fully Remote (100% Remote): ${remote_salary:,.0f}")
```

```python
import matplotlib.pyplot as plt

# Define categories and salaries
categories = [0, 50, 100]  # 0% = On-Site, 50% = Hybrid, 100% = Fully Remote
salaries = [on_site_salary, hybrid_salary, remote_salary]

# Create the bar chart
plt.figure(figsize=(8, 5))
plt.bar(categories, salaries, color=["green", "red", "orange"], width=20)

# Add title and labels
plt.title("Average Salary by Remote Work Ratio", fontsize=14, fontweight="bold")
plt.xlabel("Remote Work Categories", fontsize=12)
plt.ylabel("Average Salary (USD)", fontsize=12)

# Set correct X-axis labels
plt.xticks(categories, labels=["On-Site", "Hybrid", "Remote"], rotation=80, fontsize=10)
plt.yticks(fontsize=10)

# Show the chart
plt.show()
```
